```python
import numpy as np

class Steepest_DescentLinearRegression:
    def __init__(self, alpha=0.0001, _epochs=1000):
        self.alpha, self._epochs = alpha, _epochs
    
    def fit(self, X, y):
        b = 0
        m = 5
        n = X.shape[0]
        for _ in range(self._epochs):
            b_gradient = -2 * np.sum(y - m*X + b) / n 
            m_gradient = -2 * np.sum(X*(y - (m*X + b))) / n
            b = b + (self.alpha * b_gradient)
            m = m - (self.alpha * m_gradient)
        self.m, self.b = m, b
        
    def predict(self, X):
        return self.m*X + self.b


clf = Steepest_DescentLinearRegression()
clf.fit(X, y)

```

```python
np.random.seed(42)
X = np.array(sorted(list(range(5))*20)) + np.random.normal(size=100, scale=0.5)
y = np.array(sorted(list(range(5))*20)) + np.random.normal(size=100, scale=0.25)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X, y, color='red')
plt.plot(X, clf.predict(X))
plt.gca().set_title("Gradient Descent Linear Regressor")
```
